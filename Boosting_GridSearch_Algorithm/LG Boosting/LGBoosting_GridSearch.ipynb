```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor
param_grid={'n_estimators':[10,50,100,1000],'learning_rate':[0.01,0.3]}
grid=GridSearchCV(LGBMRegressor(),param_grid,refit=True,verbose=2, n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_

print("The R_score value for the best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input('Age:'))
bmi_input=float(input('BMI:'))
children_input=int(input('Children:'))
sex_male_input=int(input('Sex_Male 0 or 1:'))
smoker_yes_input=int(input('Smoker_Yes 0 or 1:'))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print('Future Prediction={}'.format(Future_Prediction))
```
